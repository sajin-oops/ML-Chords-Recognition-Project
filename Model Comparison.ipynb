```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

model_results = {
    "Model":[
        "Random Forest",
        "SVM",
        "KNN",
        "Logistic Regression"
    ],
    "Accuracy":[
        0.9475409836065574,
        0.9442622950819672,
        0.8918032786885246,
        0.8754098360655738
    ]
}
df_results = pd.DataFrame(model_results)
df_results
```

```python
df_results_sorted = df_results.sort_values(by = "Accuracy", ascending = False)
df_results_sorted
```

```python
plt.figure(figsize = (8,5))
colors = plt.cm.Set2(range(len(df_results_sorted)))
bars = plt.bar(
    df_results_sorted["Model"],
    df_results_sorted["Accuracy"],
    color = colors
)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() /2,
        height + 0.002,
        f"{height:3f}",
        ha='center',
        va='bottom',
        fontsize = 9
    )

plt.xlabel("Machine Learning Model")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.ylim(0.8,1.0)
plt.xticks(rotation = 20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(
    df_results_sorted["Model"],
    df_results_sorted["Accuracy"],
    marker = 'o'
)
plt.xlabel("Machine Learning Model")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison Across Models")
plt.ylim(0.8,1.0)
plt.show()
```
