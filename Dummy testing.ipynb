```python
import os
from pydub import AudioSegment
```

```python
input_folder = "voice_memos"
output_folder = "wav_files" 

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.endswith(".m4a"):
        audio = AudioSegment.from_file(os.path.join(input_folder, file), format="m4a")
        audio.export(os.path.join(output_folder, file.replace(".m4a", ".wav")), format="wav")
```
